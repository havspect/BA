```python
import pandas as pd
import configparser
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%aimport helper
%aimport import_dfs

from IPython.core.display import display, HTML
from matplotlib.pyplot import figure
from import_dfs import Datenbank
import helper
```

```python
'''Datentypen
    staedte: Stadtdaten
    technologiedaten: Technologiedaten
    verbesserungsfaktoren: Verbesserungsfaktoren
    rf_pv_basis: Referenzstrekcen Basis
    rf_pv_pod: Referenzstrecken Pod
    rf_gv_basis: Referenzstrecken Güterverkehr Basis
    rf_gv_pod: Referenzstrecken Güterverkehr Pod
    rf_pv_Strecken: Strecken Personenverkehr 
    rf_gv_Strecken: Strecken Güterverkehr
    gemis_data: Daten zur Umrechnung Liter in Emissionen
    spezifischer_verbrauch: Je Klasse der Straßenverkehrsmittel
    bevoelkerungs_und_emissiondaten: BEV und Emissionsdaten
'''

db = Datenbank()
```

```python
tech_data = db.technologiedaten
cities = db.staedte.drop(columns=['stadt_typ', 'stadt_größe'])
cities.head()
```

```python
rf_pv_basis = db.rf_pv_basis.copy()
rf_pv_basis.head()
```

```python
rf_pv_routes = db.rf_pv_strecken.drop(columns='strecken_typ').assign(stadt_id_ende=lambda x: x['stadt_id_ende'].astype(int),
                                                                    stadt_id_start=lambda x: x['stadt_id_start'].astype(int))
rf_pv_routes.head()
```

```python
df = (rf_pv_basis.merge(rf_pv_routes, left_on='name_verbindung', right_index=True)
             .merge(cities, left_on='stadt_id_start', right_on='stadt_id', suffixes=(False, '_start'))
            .merge(cities, left_on='stadt_id_ende', right_on='stadt_id', suffixes=(False, '_ende')))
df = df.drop(columns=['stadt_id_ende', 'stadt_id_start', 'stadt_idFalse', 'stadt_nameFalse', 'stadt_name_ende', 'luftlinie'])
df.columns
```

```python
df = df.copy()

unique_modes = [x for x in helper.find_unique_modes(tech_data) if x in df.columns]
unique_modes_time = [f'{x}_time' for x in helper.find_unique_modes(tech_data) if x in df.columns]
empty_df = pd.DataFrame(0, columns=unique_modes_time, index=df.index)
df = pd.concat([df, empty_df], axis=1)
```

```python
result = [None]*(len(unique_modes)+len(unique_modes_time))
result[::2] = unique_modes
result[1::2] = unique_modes_time


df = df.reindex(['strecken_typ', 'name_verbindung', 'transferzeit', 'wartezeit', 'fahrtzeit', 'length'] 
                + result
                + ['stadt_start', 'stadt_ende', 'startpunkt', 'endpunkt'] ,axis='columns'). sort_values(by=['strecken_typ', 'name_verbindung']).drop(columns = ['bike_sharing' ,'bike_sharing_time'])
```

```python
df.to_excel('verarbeitung.xlsx')
```
